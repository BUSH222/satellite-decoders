```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
```

```python
sample_rate = 48000
sample_interval = 1.0/sample_rate
T = 4*np.pi
t = np.arange(0,T, sample_interval)

freq = 1
DC = 0.5
# sine: Asin(kx+phi)+DC, where A is the amplitude, k is the frequency in Hz, phi is the phase shift in radians
x_1 = 3*np.sin(t + 0)
x_2 = np.sin(4*t + 0)
x_3 = 0.5*np.sin(7*t + 0)

x = DC + x_1 + x_2 + x_3
plt.plot(t, x, 'r')
plt.show()
```

```python
amplitudes = [(abs(c)*2).round(2) for c in ifft(x)]
phases = [(np.angle(c)).round(2) for c in ifft(x)]
DC = amplitudes[0]/2
print(phases[0:100])
```

```python
def restore_function(t_val):
    return DC + sum([amplitudes[i] * np.sin(i * t_val / (T/(2*np.pi)) + 0) for i in range(len(amplitudes[1:100]))])

plt.plot(t, restore_function(t), 'r')
plt.show()
```
