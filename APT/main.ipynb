```python
import numpy as np
import matplotlib.pyplot as plt
import pydub
import os
from scipy import iftt
```

```python
F = os.path.join(os.path.abspath(""), 'source.mp3')
a = pydub.AudioSegment.from_mp3(F)
raw_data = np.array(a.get_array_of_samples())

print(a.frame_rate)
print(a.channels)
print(a.sample_width)
print(a.duration_seconds)
print(a.frame_count())
print(len(raw_data))
print(a.frame_rate * a.duration_seconds)
```

```python
plt.plot(raw_data[0:a.frame_rate])  # 1 second of audio
plt.show()  # You can see 2 identical patterns - 2 APT frames
```
