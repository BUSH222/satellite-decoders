```python
with open('jpss_tlm.cadu', 'rb') as f:
    data = f.read()
```

```python
# locate cadu markers 

marker = b'\x1A\xCF\xFC\x1D'
marker_bits = ''.join(f'{byte:08b}' for byte in marker)
print(marker_bits)
positions = [i for i in range(len(data)) if data[i:i+4] == marker]
print(positions)
positions_differences = [j - i for i, j in zip(positions[:-1], positions[1:])]
print(positions_differences)

# each cadu block is 1279 bytes long
cadu_blocks = [data[i:i+1279] for i in positions]
```
